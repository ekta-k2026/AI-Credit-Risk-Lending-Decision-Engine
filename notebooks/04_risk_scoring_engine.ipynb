```python
import pandas as pd
import joblib

model = joblib.load("../models/logistic_regression_model.pkl")
scaler = joblib.load("../models/scaler.pkl")

print("Model Loaded Successfully")
```

```python
df = pd.read_csv(
    "../data/processed/credit_risk_model_data.csv"
)

print(df.shape)
```

```python
X = df.drop("default", axis=1)

y = df["default"]

print(X.shape)
print(y.shape)
```

```python
X_scaled = scaler.transform(X)

print(X_scaled.shape)
```

```python
probabilities = model.predict_proba(X_scaled)

print(probabilities[:5])
```

```python
default_probability = probabilities[:, 1]

print(default_probability[:10])
```

```python
risk_score = (default_probability * 100).round(0)

risk_score[:10]
```

```python
df["risk_score"] = risk_score

df[
    [
        "default",
        "risk_score"
    ]
].head()
```

```python
def risk_category(score):

    if score <= 30:
        return "Low Risk"

    elif score <= 70:
        return "Medium Risk"

    else:
        return "High Risk"
```

```python
df["risk_category"] = df["risk_score"].apply(
    risk_category
)

df["risk_category"].value_counts()
```

```python
def lending_decision(category):

    if category == "Low Risk":
        return "Approve"

    elif category == "Medium Risk":
        return "Review"

    else:
        return "Reject"
```

```python
df["decision"] = df["risk_category"].apply(
    lending_decision
)

df["decision"].value_counts()
```

```python
df[
    [
        "default",
        "risk_score",
        "risk_category",
        "decision"
    ]
].head(20)
```

```python
df["risk_score"].describe()
```

```python
def risk_category_v2(score):

    if score <= 20:
        return "Low Risk"

    elif score <= 40:
        return "Medium Risk"

    else:
        return "High Risk"
```

```python
df["risk_category_v2"] = df["risk_score"].apply(
    risk_category_v2
)

df["risk_category_v2"].value_counts()
```

```python
def lending_decision_v2(category):

    if category == "Low Risk":
        return "Approve"

    elif category == "Medium Risk":
        return "Review"

    else:
        return "Reject"
```

```python
df["decision_v2"] = df["risk_category_v2"].apply(
    lending_decision_v2
)

df["decision_v2"].value_counts()
```

```python
df.to_csv(
    "../data/processed/credit_risk_scored.csv",
    index=False
)

print("Scored dataset saved successfully")
```
