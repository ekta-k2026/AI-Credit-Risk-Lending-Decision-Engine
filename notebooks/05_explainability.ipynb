```python
def explain_applicant(row):

    reasons = []

    if row["int_rate"] > 15:
        reasons.append("High Interest Rate")

    if row["dti"] > 20:
        reasons.append("High Debt-to-Income Ratio")

    if row["avg_fico"] < 680:
        reasons.append("Low FICO Score")

    if row["annual_inc"] < 50000:
        reasons.append("Lower Annual Income")

    if row["loan_amnt"] > 15000:
        reasons.append("Large Loan Amount")

    if len(reasons) == 0:
        reasons.append("Strong Overall Credit Profile")

    return reasons
```

```python
df["risk_reasons"] = df.apply(
    explain_applicant,
    axis=1
)
```
